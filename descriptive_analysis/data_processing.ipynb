```python
# Run this block of code to process data, import the necessary libraries and functions

def handle_encoding(encoding):
    # if there is no , between ] and [ then add one
    if '][' in str(encoding):
        encoding = str(encoding).replace('][', '],[')
    if ' ' in str(encoding):
        encoding = str(encoding).replace(' ', '')
    # remove quotation marks
    encoding = str(encoding).replace("'", '')
    # remove square brackets
    encoding = str(encoding).replace('[', '').replace(']', '')
    # split by comma
    encoding = encoding.split(',')
    # remove spaces
    encoding = [x.strip() for x in encoding]
    # capitalize all words
    encoding = [x.upper() for x in encoding]
    return encoding

def ignore_numbers_in_encoding(encoding: list):
    encoding = [value.split('>')[0] for value in encoding]
    return encoding

def plot_heat_map(df, title, x_label, y_label):
    """
    Plot a heatmap for the given dataframe
    :param df: dataframe
    :param title: title of the heatmap
    :return: None
    """
    import seaborn as sns
    import matplotlib.pyplot as plt
    # plot heatmap
    sns.heatmap(df, annot=False, fmt="d")
    # change heatmap shape
    plt.gcf().set_size_inches(4, 3)
    # set title
    # plt.yticks(rotation=45)
    # show plot
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.savefig('output/' + title + '.pdf')

from docx import Document

doc = Document('Encoding.docx')
table = doc.tables[0]
Appl_full_name = table.cell(1,1).text.split('\n')
Appl_encoding = table.cell(1,2).text.split('\n')
appl_dict = dict(zip(Appl_encoding, Appl_full_name))

PROB_full_name = table.cell(2,1).text.split('\n')
PROB_encoding = table.cell(2,2).text.split('\n')
prob_dict = dict(zip(PROB_encoding, PROB_full_name))

CONT_full_name = table.cell(3,1).text.split('\n')
CONT_encoding = table.cell(3,2).text.split('\n')
cont_dict = dict(zip(CONT_encoding, CONT_full_name))

AIME_full_name = table.cell(4,1).text.split('\n')
AIME_encoding = table.cell(4,2).text.split('\n')
aime_dict = dict(zip(AIME_encoding, AIME_full_name))

SECU_full_name = table.cell(5,1).text.split('\n')
SECU_encoding = table.cell(5,2).text.split('\n')
secu_dict = dict(zip(SECU_encoding, SECU_full_name))

PRIV_full_name = table.cell(6,1).text.split('\n')
PRIV_encoding = table.cell(6,2).text.split('\n')
priv_dict = dict(zip(PRIV_encoding, PRIV_full_name))

ALLO_full_name = table.cell(7,1).text.split('\n')
ALLO_encoding = table.cell(7,2).text.split('\n')
allo_dict = dict(zip(ALLO_encoding, ALLO_full_name))

METR_full_name = table.cell(8,1).text.split('\n')
METR_encoding = table.cell(8,2).text.split('\n')
metr_dict = dict(zip(METR_encoding, METR_full_name))

SUST_full_name = table.cell(10,1).text.split('\n')
SUST_encoding = table.cell(10,2).text.split('\n')
sust_dict = dict(zip(SUST_encoding, SUST_full_name))

EVAL_full_name = table.cell(12,1).text.split('\n')
EVAL_encoding = table.cell(12,2).text.split('\n')
eval_dict = dict(zip(EVAL_encoding, EVAL_full_name))

BLOC_full_name = table.cell(13,1).text.split('\n')
BLOC_encoding = table.cell(13,2).text.split('\n')
bloc_dict = dict(zip(BLOC_encoding, BLOC_full_name))

CONS_full_name = table.cell(14,1).text.split('\n')
CONS_encoding = table.cell(14,2).text.split('\n')
cons_dict = dict(zip(CONS_encoding, CONS_full_name))

TECH_full_name = table.cell(9,1).text.split('\n')
TECH_encoding = table.cell(9,2).text.split('\n')
tech_dict = dict(zip(TECH_encoding, TECH_full_name))

COMM_full_name = table.cell(11,1).text.split('\n')
COMM_encoding = table.cell(11,2).text.split('\n')
comm_dict = dict(zip(COMM_encoding, COMM_full_name))

import matplotlib
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 13}
matplotlib.rc('font', **font)

import pandas as pd
df = pd.read_excel('../ReviewedRecords.xlsx', sheet_name='ExtractedFeatures')
# handle encoding for D2: APPL [CODING]
df['D2: APPL [CODING]'] = df['D2: APPL [CODING]'].apply(lambda x: handle_encoding(x))
# handle encoding for D3: PROB [CODING]
df['D3: PROB [CODING]'] = df['D3: PROB [CODING]'].apply(lambda x: handle_encoding(x))
# handle encoding for D4: CONT [CODING]
df['D4: CONT [CODING]'] = df['D4: CONT [CODING]'].apply(lambda x: handle_encoding(x))
# handle encoding for D5: AIME [CODING>EDGE|BLOC|BOTH]
df['D5: AIME [CODING>EDGE|BLOC|BOTH]'] = df['D5: AIME [CODING>EDGE|BLOC|BOTH]'].apply(lambda x: handle_encoding(x))
# handle encoding for D6: SECU [CODING>EDGE|BLOC|BOTH]
df['D6: SECU [CODING>EDGE|BLOC|BOTH]'] = df['D6: SECU [CODING>EDGE|BLOC|BOTH]'].apply(lambda x: handle_encoding(x))
# handle encoding for D7: PRIV [CODING>EDGE|BLOC|BOTH]
df['D7: PRIV [CODING>EDGE|BLOC|BOTH]'] = df['D7: PRIV [CODING>EDGE|BLOC|BOTH]'].apply(lambda x: handle_encoding(x))
# handle encoding for D8: ALLO [CODING>EDGE|BLOC|BOTH]
df['D8: ALLO [CODING>EDGE|BLOC|BOTH]'] = df['D8: ALLO [CODING>EDGE|BLOC|BOTH]'].apply(lambda x: handle_encoding(x))
# handle encoding for D9: METR[CODING>EDGE|BLOC|BOTH]
df['D9: METR[CODING>EDGE|BLOC|BOTH]'] = df['D9: METR[CODING>EDGE|BLOC|BOTH]'].apply(lambda x: handle_encoding(x))
# handle encoding for D11: TECH [CODING>EDGE|BLOC|BOTH]
df['D11: TECH [CODING>EDGE|BLOC|BOTH]'] = df['D11: TECH [CODING>EDGE|BLOC|BOTH]'].apply(lambda x: handle_encoding(x))
# handle encoding for D15: SUST [CODING>EDGE|BLOC|BOTH]
df['D15: SUST [CODING>EDGE|BLOC|BOTH]'] = df['D15: SUST [CODING>EDGE|BLOC|BOTH]'].apply(lambda x: handle_encoding(x))
# handle encoding for D17: COMM [CODING>EDGE|BLOC|BOTH]
df['D17: COMM [CODING>EDGE|BLOC|BOTH]'] = df['D17: COMM [CODING>EDGE|BLOC|BOTH]'].apply(lambda x: handle_encoding(x))
# handle encoding for D18: EVAL [CODING]
df['D18: EVAL [CODING]'] = df['D18: EVAL [CODING]'].apply(lambda x: handle_encoding(x))
# handle encoding for D19: BLOC [CODING]
df['D19: BLOC [CODING]'] = df['D19: BLOC [CODING]'].apply(lambda x: handle_encoding(x))
# handle encoding for D20: CONS [CODING]
df['D20: CONS [CODING]'] = df['D20: CONS [CODING]'].apply(lambda x: handle_encoding(x))

# remove rows where Violated [1,2,3,4] is not empty
df = df[df['Violated [1,2,3,4]'].isnull()]

flatten_df = df.explode('D2: APPL [CODING]').explode('D3: PROB [CODING]').explode('D4: CONT [CODING]')
# remove NONE and NAN
flatten_df = flatten_df[flatten_df['D2: APPL [CODING]'] != 'NONE']
flatten_df = flatten_df[flatten_df['D2: APPL [CODING]'] != 'NAN']
flatten_df = flatten_df[flatten_df['D3: PROB [CODING]'] != 'NONE']
flatten_df = flatten_df[flatten_df['D3: PROB [CODING]'] != 'NAN']
flatten_df = flatten_df[flatten_df['D4: CONT [CODING]'] != 'NONE']
flatten_df = flatten_df[flatten_df['D4: CONT [CODING]'] != 'NAN']
# rename values in D2: APPL [CODING]
flatten_df['D2: APPL [CODING]'] = flatten_df['D2: APPL [CODING]'].replace(appl_dict)
flatten_df['D3: PROB [CODING]'] = flatten_df['D3: PROB [CODING]'].replace(prob_dict)
flatten_df['D4: CONT [CODING]'] = flatten_df['D4: CONT [CODING]'].replace(cont_dict)

# remove nan in publication year
flatten_df = flatten_df[flatten_df['Publication Year'].notna()]
flatten_df['Publication Year'] = flatten_df['Publication Year'].apply(lambda x: int(x))
flatten_df = flatten_df[flatten_df['Publication Year'].notna()]
flatten_df['Publication Year'] = flatten_df['Publication Year'].astype(int)
```
